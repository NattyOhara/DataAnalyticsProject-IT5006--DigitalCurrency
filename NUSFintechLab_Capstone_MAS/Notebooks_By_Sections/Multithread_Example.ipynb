### Import libraries

```python
#libraries for general purpose
import threading
import time
```

## Multi-thread - simple example

```python
#base functions
def boil_noodle(seconds):
  print('Boiling noodle for {} secondes\n'.format(seconds))
  time.sleep(seconds)
  print('Noodle is boiled \n')

def prepare_soup(seconds):
  print('Preparing soup for the noodle for {} seconds \n'.format(seconds))
  time.sleep(seconds)
  print('Soup is prepared \n')
```

### No multi thread

Without multi-threading, each functions are operated one by one, such that the noodle can get soggy and not tasty!

```python
print('Start cooking noodle')
boil_noodle(3)
prepare_soup(2)
print('Put noodle and soup together in the bowl')
print('Ramen is done!')
```

### Multi thread

By using threading, we can see that boiling noodle and preparing soup are executed concurrently, such that we can enjoy tasty noodle! 

```python
print('Start cooking noodle')
#Creating threads
thread1 = threading.Thread(target=boil_noodle, args=(3,))
thread2 = threading.Thread(target=prepare_soup, args=(2,))
#Starting thread 1
thread1.start()
#Starting thread 2
thread2.start()
#Wait until the execution of thread 1 is done
thread1.join()
#Wait until the execution of thread 2 is done
thread2.join()
#Both thread 1, 2 are done
print('Put noodle and soup together in the bowl')
print('Ramen is done!')
```
